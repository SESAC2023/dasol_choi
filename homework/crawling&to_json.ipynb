```python
def csv_to_dict(i):
    dic = {
        "num": "",
        "name": "",
        "url": "",
        "price": "",
        "img_url": "",
        "info": [],
        "skin_type": "",
        "skin_irritation": "",
        "effect": "",
        "ingredient_tag": [],
        "star": "", 
        "reviews": [],
        "private_stars": "" 
    }
    dic['num'], dic['name'], dic['url'] = db.loc[i, 'no'], db.loc[i, 'name'], db.loc[i, 'url']
    dic['info'] = [db.loc[i, 'info1'], db.loc[i, 'info2'], db.loc[i, 'info3']]
    dic['skin_type'], dic['skin_irritation'] = db.loc[i, 'skin type'], db.loc[i, 'skin irritation']
    dic['effect'] = [db.loc[i, 'effect1'], db.loc[i, 'effect2'], db.loc[i, 'effect3'], db.loc[i, 'effect4']]
    dic['ingredient_tag'] = [db.loc[i, 'ingredent_tag1'], db.loc[i, 'ingredent_tag2'], db.loc[i, 'ingredent_tag3']]
    globals()[f'dict{i+1}'] =  dic
    return globals()[f'dict{i+1}']
```

```python
for i in range(db.shape[0]):
    csv_to_dict(i)
```

### # 올리브영 제품 정보 크롤링 (이미지, 가격, 전체 별점, 리뷰 20개, 리뷰 별점 )

```python
def get_info(url):
    # 브라우저 생성
    browser = webdriver.Chrome('./chromedriver.exe')
    time.sleep(3)
    browser.get(f'https://www.oliveyoung.co.kr/store/goods/{url}')
    time.sleep(5)
    
    # 리뷰버튼 클릭
    browser.find_element_by_css_selector('a.goods_reputation').click()
    time.sleep(3)
    
    # 이미지 url & 가격 저장
    img_url = browser.find_element_by_id('mainImg').get_attribute('src')
    price = browser.find_element_by_class_name('price-2').text

    # 전체 별점 저장 
    full_html = browser.page_source
    soup = BeautifulSoup(full_html, 'html.parser')
    star = soup.select('#repReview > b')[0].string.replace('\t', '').replace('\n', '')
    
    # 리뷰, 리뷰 별점 저장
    reviews, private_stars = [], []
    review = soup.find_all('div', class_ = 'txt_inner')
    private_star = soup.find_all('span', class_ = 'point')[1:]
    for i in range(len(review)):
        reviews.append(review[i].get_text().replace('\xa0', ' '))
        private_stars.append(private_star[i].string)       
    time.sleep(3)
    
    # 2페이지로 이동
    try:
        if len(soup.select('.pageing')[0].get_text().split('\n')) >= 3 :
            browser.find_element_by_class_name('pageing').find_element_by_tag_name('a').send_keys(Keys.ENTER)
        else:
            return img_url, price, star, reviews, private_stars
    except:
        return img_url, price, star, reviews, private_stars
    time.sleep(3)
    
    full_html = browser.page_source
    soup = BeautifulSoup(full_html, 'html.parser')
    review = soup.find_all('div', class_ = 'txt_inner')
    private_star = soup.find_all('span', class_ = 'point')[1:]
    for i in range(len(review)):
        reviews.append(review[i].get_text().replace('\xa0', ' '))
        private_stars.append(private_star[i].string) 
        
    return img_url, price, star, reviews, private_stars

# 크롤링 한 정보들 dictionary 변수에 추가
def save_info(index): 
    img_url, price, star, reviews, private_stars = get_info(urls[index])
    globals()[f'dict{index+1}']['star'] = star
    globals()[f'dict{index+1}']['price'] = price
    globals()[f'dict{index+1}']['img_url'] = img_url
    globals()[f'dict{index+1}']['private_stars'] = private_stars
    globals()[f'dict{index+1}']['reviews'] = reviews
    return globals()[f'dict{index+1}']
```

```python
# db에 있는 전체 상품에 대해 크롤링 
for i in range(db.shape[0]):  
    globals()[f'dict{i+1}'] = save_info(i)
    print(f'dict{i+1} success!')
print('------------------')
print('all success!')
```

### # dictionary to json

```python
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8, np.int16, np.int32, 
                            np.int64, np.uint8, np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, np.float64)):
            return float(obj)
        elif isinstance(obj, (np.complex_, np.complex64, np.complex128)):
            return {'real': obj.real, 'imag': obj.imag}
        elif isinstance(obj, (np.ndarray,)):
            return obj.tolist()
        elif isinstance(obj, (np.bool_)):
            return bool(obj)
        elif isinstance(obj, (np.void)): 
            return None
        return json.JSONEncoder.default(self, obj)
```

```python
import json
def write_to_json(dictionary_data):
    with open('db.json', 'w', encoding='UTF8') as file:
        json.dump(dictionary_data, file, cls=NumpyEncoder, ensure_ascii=False)
write_to_json(all_dicts)
```

```python
def get_name(db):
    if db['name'][i].split(' ')[-1] == '':
        name = ' '.join(db['name'][i].split(' ')[:-2])
    else:
        name = ' '.join(db['name'][i].split(' ')[:-1])
    return name
```

```python
def scroll_down():
    #스크롤 내리기
    last_page_height = browser.execute_script("return document.documentElement.scrollHeight")
    browser.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1)
    new_page_height = browser.execute_script("return document.documentElement.scrollHeight")
    if new_page_height == last_page_height:
        time.sleep(1)
    else:
        last_page_height = new_page_height
```

```python
from urllib.request import urlopen

def save_images(n):
    full_html = browser.page_source
    time.sleep(2)
    soup = BeautifulSoup(full_html, 'lxml')
    images = soup.find_all('img', {'class' : '_image _listImage'})
    for idx, image in enumerate(images):
        if image['src'].startswith('https://'):
            img_url = image['src']
        elif image['data-lazy-src'].startswith('data'):
            img_url = image['data-lazy-src']
        else:
            continue
        with urlopen(img_url) as f:
            with open(f'D:/olive_young/naver_product_img/{i+1}/{n}.jpg','wb') as h: # w - write b - binary
                img = f.read()
                h.write(img)
                n += 1
```
